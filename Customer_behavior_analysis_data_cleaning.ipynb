##### _Handle the null Values_

_Size_

```python
df['Size'].unique()
```

```python
df.head()
```

```python
df.groupby('Category')['Size'].unique()
```

```python
df.groupby('Item Purchased')['Size'].unique()
```

_Treat the null value or replacing all values is the Electronics category_

```python
df.loc[df['Category'] == 'Electronics', 'Size'] = 'Not Applicable'
```

```python
df.groupby('Category')['Size'].unique()
```

_Treat the null value or replacing all values is the Accessories category_

```python
df.loc[df['Category'] == 'Accessories', 'Size'] = 'Free Size'
```

```python
df.groupby('Category')['Size'].unique()
```

_Treat the null value or replacing all values is the Footwear category_

```python
df.loc[df['Category'] == 'Footwear', 'Size'] = 'Not Available'
```

```python
df.groupby('Category')['Size'].unique()
```

_Treat null values or replacing all vlaues in the Clothing category_

```python
df.loc[
    (df['Category'] == "Clothing") & (df["Size"].isnull()), 
    "Size"
] = df[df["Category"] == "Clothing"]["Size"].mode()[0]
```

```python
df.groupby('Category')['Size'].unique()
```

_Filling null values of Review Ratings with the mean of each Item Purchased_

```python
df['Review Rating'] = df['Review Rating'].fillna(
    df.groupby('Item Purchased')['Review Rating'].transform('mean')
)
```
